```python
import pandas as pd
```

```python
#List of CSV file to combined
csv_files = [
    'Page_01.csv',
    'Page_02_05.csv',
    'Page_06_10.csv',
    'Page_11_15.csv',
    'Page_16_20.csv',
    'Page_21_25.csv',
    'Page_26_30.csv'
]

#Read and combine CSV files into a DataFrame
combined_df = pd.concat([pd.read_csv(file) for file in csv_files])

#Convert to "DienThoai.csv" file
combined_df.to_csv("DienThoai.csv", index=False)
```
