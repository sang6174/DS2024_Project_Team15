```python
import numpy as np
from selenium import webdriver
from time import sleep
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException, ElementClickInterceptedException
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import tqdm
import funcs
```

```python
driver = funcs.initializeChrome()
link = "https://www.lazada.vn/catalog/?page=69&q=gi%C3%A0y%20d%C3%A9p&spm=a2o4n.home-vn.search.d_go"
driver.get(link)
```

```python
link = "https://www.lazada.vn/catalog/?page=69&q=gi%C3%A0y%20d%C3%A9p&spm=a2o4n.home-vn.search.d_go"
dataProducts = funcs.getDataInNPage(driver, link, 3)
```

```python
df = pd.DataFrame(dataProducts)
```

```python
detailProducts = {
    "Original": [],
    "Ship_price": [],
    "Return": [],
    "Positive_review": [],
    "Ship_on_time": [],
    "Rate_response": [],
    "One_star": [],
    "Two_star": [],
    "Three_star": [],
    "Four_star": [],
    "Five_star": []
}
```

```python
# Create a progress bar to track the crawl process
for link in tqdm(df.Links, desc="Crawling product details"):
    details = funcs.getDetailItems(driver, link)
    for key in detailProducts.keys():
        detailProducts[key].append(details[key])
```

```python
# add columns to df
for key, value in detailProducts.items():
    df[key] = value
```

```python
driver.close()
```

```python
# save to csv
df.to_csv("lazada1_9.csv", index = True)
df
```
