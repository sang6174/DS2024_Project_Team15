```python
import pandas as pd
import numpy as np
```

```python
#List of CSV file to combined
csv_files = [
    #Thoi trang
    'quanaollms.csv',
    'giaydepllms.csv',
    'Data_Jewelry_After_LLMs.csv',

    #Dien tu
    'LLM_DienThoai.csv',
    'Data_Laptop_After_LLMs.csv',
    'LLM_Tivi.csv'
]

#Read and combine CSV files into a DataFrame
data_final = pd.concat([pd.read_csv(file) for file in csv_files])
#Reset index
data_final.reset_index(drop=True, inplace=True)
data_final.index = np.arange(1, len(data_final)+1)


#Convert to "DienThoai.csv" file
data_final.to_csv("Final_Data_LLM.csv", index=False)
```
